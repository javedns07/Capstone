```python
#importing libraries
import selenium
import pandas as pd
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
```

```python
# empty lists
Sno = []
BHK_type = []
Property_Type = []
beds = []
bathroom = []
Balcony = []
Address = []
City = []
State_code = []
Country = []
Landmarks = []
carpetarea = []
status = []
floor = []
transaction_type = []
Year_Built = []
price_per_square_feet = []
furnished_status = []
facing = []
Car_Parking = []
Type_of_ownership = []
booking_amount = []
buy_total_price = []

# chrome browser to webdriver
driver=webdriver.Chrome()
```

```python
# Load website
driver.get("https://vestapreferred.com/")
```

```python
#maximizing the window
driver.maximize_window()
```

```python
# Find search box 
search_box = driver.find_element(By.XPATH,"/html/body/main/header/div/div[2]")
search_box.click()
```

```python
# Find search box 
search_box = driver.find_element(By.XPATH,"/html/body/main/div[1]/div[2]/ul/li[1]/a")
search_box.click()
```

```python
propertiess=[]
```

```python
for p in driver.find_elements(By.XPATH,'//div[@itemtype="http://schema.org/House"]/a'):
    propertiess.append(p.get_attribute('href'))
```

```python
for i in propertiess:  # Iterating each and every URL to fetch full details
    driver.get(i)  # Taking each URL one by one as an input
    sleep(1)

    # BHK_type 
    try:
        bhk = driver.find_element(By.XPATH, '//div[@class="flexbox hidden-mobile"]/div[2]/div[1]')
        BHK_type.append(bhk.text)
    except:
        BHK_type.append('None')

    # BEDS 
    try:
        beds_ = driver.find_element(By.XPATH, '//div[@class="flexbox hidden-mobile"]/div[2]/div[1]')
        beds.append(beds_.text)
    except:
        beds.append('None')

    # Bathroom
    try:
        bath = driver.find_element(By.XPATH, '//div[@class="flexbox hidden-mobile"]/div[3]/div[1]')
        bathroom.append(bath.text)
    except:
        bathroom.append('None')

    # Carpet area
    try:
        cart_area = driver.find_element(By.XPATH, '//div[@class="flexbox hidden-mobile"]/div[4]/div[1]')
        carpetarea.append(cart_area.text)
    except:
        carpetarea.append('None')

    # Price per square feet
    try:
        PRSF = driver.find_element(By.XPATH, '//div[@class="flexbox hidden-mobile"]/div[5]/div[1]')
        price_per_square_feet.append(PRSF.text)
    except:
        price_per_square_feet.append('None')

    # Property_Type
    try:
        Pro_ = driver.find_element(By.XPATH, '//div[@class="outer-parameters-block"]/ul/li[2]/div[2]')
        Property_Type.append(Pro_.text)
    except:
        Property_Type.append('None')

    # Status
    try:
        sta = driver.find_element(By.XPATH, '//div[@class="outer-parameters-block"]/ul/li[8]/div[2]')
        status.append(sta.text if sta.text == 'Active' else 'None')
    except:
        status.append('None')

    # Year Built
    try:
        Year = driver.find_element(By.XPATH, '//div[@class="outer-parameters-block"]/ul/li[10]/div[2]')
        year_text = Year.text.strip()
        
        if year_text.isdigit() and len(year_text) == 4:
            Year_Built.append(int(year_text))
        else:
            Year_Built.append('None')  # Append 'NaN' if not a valid year
    except:
        Year_Built.append('None')


    # Landmarks
    try:
        Landma = driver.find_element(By.XPATH, '//div[@class="outer-parameters-block"]/ul/li[5]/div[2]')
        Landmarks.append(Landma.text)
    except:
        Landmarks.append('None')

    # Balcony
    try:
        Balc = driver.find_element(By.XPATH, '//div[@data-testid="property-features-block"]/div[2]/ul/li[1]')
        Balcony.append(Balc.text if Balc.text == 'Balcony' else 'None')
    except:
        Balcony.append('None')

    # Address
    try:
        Addr = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[1]')
        Address.append(Addr.text)
    except:
        Address.append('None')

    # City
    try:
        Cit = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[2]')
        city_text = Cit.text.split(',')[0]  # Split the text by comma and get the first part
        City.append(city_text)
    except:
        City.append('None')

    # State_code
    try:
        State_cod = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[2]')
        State_code_ = State_cod.text.split(',')[1].strip()  # Split the text by comma and get the second part
        State_code.append(State_code_)
    except:
        State_code.append('None')

    # Country
    try:
        Addr = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        Country.append(Addr.text)
    except:
        Country.append('United States')

    # Floor
    try:
        fl = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        floor.append(fl.text)
    except:
        floor.append('None')

    # Transaction type
    try:
        trans = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        transaction_type.append(trans.text)
    except:
        transaction_type.append('None')

    # Furnished status
    try:
        furnish = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        furnished_status.append(furnish.text)
    except:
        furnished_status.append('None')

    # Facing
    try:
        fac = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        facing.append(fac.text)
    except:
        facing.append('None')

    # Car Parking
    try:
        Car_Par = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        Car_Parking.append(Car_Par.text)
    except:
        Car_Parking.append('None')

    # Type of ownership
    try:
        Type_of_owners = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        Type_of_ownership.append(Type_of_owners.text)
    except:
        Type_of_ownership.append('None')

    # Booking amount
    try:
        booking = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        booking_amount.append(booking.text)
    except:
        booking_amount.append('None')

    # Buy total price
    try:
        buy_t = driver.find_element(By.XPATH, '//div[@class="flexbox hidden-mobile"]/div[1]/div[2]')
        price_usd = buy_t.text.replace('$', '').replace(',', '')  # Remove '$' and ','
        price_inr = round(float(price_usd) * 83.12)  # Convert to float and multiply by conversion rate
        buy_total_price.append(f'{price_inr:}')  # Format the output as INR
    except:
        buy_total_price.append('None')
        
        
    
# Quit WebDriver
driver.quit()

```

```python
df1=pd.DataFrame({
    "BHK_type": BHK_type,
    "Beds": beds,
    "Bathroom": bathroom,
    "Carpet_area": carpetarea,
    "Price_per_square_feet": price_per_square_feet,
    "Property_Type": Property_Type,
    "Status": status,
    "Year_Built": Year_Built,
    "Landmarks": Landmarks,
    "Balcony": Balcony,
    "Address": Address,
    "City": City,
    "State_Code": State_code,
    "Country": Country,
    "Floor": floor,
    "Transaction_Type": transaction_type,
    "Furnished_Status": furnished_status,
    "Facing": facing,
    "Car_Parking": Car_Parking,
    "Type_of_Ownership": Type_of_ownership,
    "Booking_Amount": booking_amount,
    "Buy_Total_Price": buy_total_price
})

```

```python
#importing libraries
import selenium
import pandas as pd
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
```

```python
# empty lists
Sno = []
BHK_type = []
Property_Type = []
beds = []
bathroom = []
Balcony = []
Address = []
City = []
State_code = []
Country = []
Landmarks = []
carpetarea = []
status = []
floor = []
transaction_type = []
Year_Built = []
price_per_square_feet = []
furnished_status = []
facing = []
Car_Parking = []
Type_of_ownership = []
booking_amount = []
buy_total_price = []

# chrome browser to webdriver
driver=webdriver.Chrome()
```

```python
# Load website
driver.get("https://www.nobroker.in/")
```

```python
#maximizing the window
driver.maximize_window()
```

```python
# Find search box 
search_box = driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[1]/div[4]/button")
search_box.click()
```

```python
propertiess=[]
```

```python
for p in driver.find_elements(By.XPATH,'//h2[@class="heading-6 flex items-center font-semi-bold m-0"]/a'):
    propertiess.append(p.get_attribute('href'))
```

```python
len(propertiess)
```

```python
for i in propertiess:  # Iterating each and every URL to fetch full details
    driver.get(i)  # Taking each URL one by one as an input
    sleep(1)

    # BHK_type 
    try:
        beds_ = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[1]/div[2]/span/h4')
        bed_text = beds_.text.replace('Bedroom', '').strip()
        BHK_type.append(bed_text)
    except:
        BHK_type.append('None')

    # Bathroom
    try:
        bath = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[3]/div[2]/span/h4')
        bath_text = bath.text.replace('Bathroom', '').strip()
        bathroom.append(bath_text)
    except:
        bathroom.append('None')

    # Carpet area
    try:
        cart_area = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[6]/div[3]/h5')
        if 'Sq.Ft' in cart_area.text:
            carpet_area = cart_area.text.replace(' Sq.Ft', '').replace(',', '')  # Remove ' Sq.Ft' and commas
            carpetarea.append(carpet_area)
        else:
            carpetarea.append('None')
    except:
        carpetarea.append('None')

    # Price per square feet
    try:
        PRSF = driver.find_element(By.XPATH, '//div[@class="flexbox hidden-mobile"]/div[5]/div[1]')
        price_per_square_feet.append(PRSF.text)
    except:
        price_per_square_feet.append('None')

    # Property_Type
    try:
        Pro_ = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[6]/div[2]/span/h4')
        Property_Type.append(Pro_.text)
    except:
        Property_Type.append('None')

    # Status
    try:
        sta = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[4]/div[2]/span/h4')
        status.append(sta.text)
    except:
        status.append('None')

    # Year Built
    try:
        Year = driver.find_element(By.XPATH, '//div[@class="outer-parameters-block"]/ul/li[10]/div[2]')
        year_text = Year.text.strip()
        
        if year_text.isdigit() and len(year_text) == 4:
            Year_Built.append(int(year_text))
        else:
            Year_Built.append('None')  # Append 'None' if not a valid year
    except:
        Year_Built.append('None')

    # Landmarks
    try:
        Landma = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[3]/div/div/span[2]/a[1]')
        Landmarks.append(Landma.text)
    except:
        Landmarks.append('None')

    # Balcony
    try:
        Balc = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[5]/div[2]/span/h4')
        if Balc.text.strip().upper() == 'NA':
            Balcony.append('None')
        else:
            Balcony.append(Balc.text.strip())
    except:
        Balcony.append('None')

    # Address
    try:
        Addr = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[1]/div/div/div[1]/div[2]/div[1]/h5')
        Address.append(Addr.text)
    except:
        Address.append('None')

    # City
    try:
        Cit = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[2]')
        city_text = Cit.text.split(',')[0]  # Split the text by comma and get the first part
        City.append(city_text)
    except:
        City.append('Bengaluru')  # Default city assumed as 'Bengaluru'

    # State_code
    try:
        State_cod = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[2]')
        State_code_ = State_cod.text.split(',')[1].strip()  # Split the text by comma and get the second part
        State_code.append(State_code_)
    except:
        State_code.append('KA')  # Default state code assumed as 'KA'

    # Country
    try:
        Addr = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[2]')
        Country.append(Addr.text)
    except:
        Country.append('India')  # Default country assumed as 'India'

    # Floor
    try:
        fl = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[9]/div[3]/h5')
        if '/' in fl.text:
            floor.append(fl.text)
        else:
            floor.append('None')
    except:
        floor.append('None')

    # Transaction type
    try:
        trans = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        transaction_type.append(trans.text)
    except:
        transaction_type.append('None')

    # Furnished status
    try:
        furnish = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[7]/div[3]/h5')
        if 'Fully Furnished' in furnish.text or 'Semi' in furnish.text:
            furnished_status.append(furnish.text)
        else:
            furnished_status.append('None')
    except:
        furnished_status.append('None')

    # Facing
    try:
        fac = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[8]/div[3]/h5')
        fac_text = fac.text.lower()  # Convert text to lowercase for case insensitivity
    
        if 'north-east' in fac_text or 'northeast' in fac_text:
            facing.append('northeast')
        elif 'south-east' in fac_text or 'southeast' in fac_text:
            facing.append('southeast')
        elif 'south-west' in fac_text or 'southwest' in fac_text:
            facing.append('southwest')
        elif 'north-west' in fac_text or 'northwest' in fac_text:
            facing.append('northwest')
        elif 'north' in fac_text:
            facing.append('North')
        elif 'east' in fac_text:
            facing.append('East')
        elif 'south' in fac_text:
            facing.append('South')
        elif 'west' in fac_text:
            facing.append('West')
        else:
            facing.append('None')
    except:
        facing.append('None')

    # Car Parking
    try:
        Car_Par = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[7]/div[2]/span/h4')
        Car_Parking.append(Car_Par.text)
    except:
        Car_Parking.append('None')

    # Type of ownership
    try:
        Type_of_owners = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[2]/div[3]/h5')
        Type_of_ownership.append(Type_of_owners.text)
    except:
        Type_of_ownership.append('None')

    # Booking amount
    try:
        booking = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        booking_amount.append(booking.text)
    except:
        booking_amount.append('None')

    
    # Buy total price
    CRORE_TO_THOUSAND = 10000000
    LAC_TO_THOUSAND = 100000

    def convert_price_to_thousands(price_text):
        if 'Crores' in price_text:
            price_value = float(price_text.replace('₹', '').replace('Crores', '').strip())
            price_in_thousands = price_value * CRORE_TO_THOUSAND
        elif 'Lacs' in price_text:
            price_value = float(price_text.replace('₹', '').replace('Lacs', '').strip())
            price_in_thousands = price_value * LAC_TO_THOUSAND
        else:
            price_in_thousands = float(price_text.replace('₹', '').strip())
        return price_in_thousands

    try:
        buy_t = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[1]/div/div/div[1]/div[3]/div[1]/div/div[1]/span')
        price_text = buy_t.text
        price_in_thousands = convert_price_to_thousands(price_text)
        buy_total_price.append(f'{round(price_in_thousands)}') 
    except:
        buy_total_price.append('None')
    
# Quit WebDriver
driver.quit()

```

```python
for i in propertiess:  # Iterating each and every URL to fetch full details
    driver.get(i)  # Taking each URL one by one as an input
    sleep(1)

    # BHK_type 
    try:
        beds_ = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[1]/div[2]/span/h4')
        bed_text = beds_.text.replace('Bedroom', '').strip()
        BHK_type.append(bed_text)
    except:
        BHK_type.append('None')

    # Bathroom
    try:
        bath = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[3]/div[2]/span/h4')
        bath_text = bath.text.replace('Bathroom', '').strip()
        bathroom.append(bath_text)
    except:
        bathroom.append('None')

    # Carpet area
    try:
        cart_area = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[6]/div[3]/h5')
        if 'Sq.Ft' in cart_area.text:
            carpet_area = cart_area.text.replace(' Sq.Ft', '').replace(',', '')  # Remove ' Sq.Ft' and commas
            carpetarea.append(carpet_area)
        else:
            carpetarea.append('None')
    except:
        carpetarea.append('None')

    # Price per square feet
    try:
        PRSF = driver.find_element(By.XPATH, '//div[@class="flexbox hidden-mobile"]/div[5]/div[1]')
        price_per_square_feet.append(PRSF.text)
    except:
        price_per_square_feet.append('None')

    # Property_Type
    try:
        Pro_ = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[6]/div[2]/span/h4')
        Property_Type.append(Pro_.text)
    except:
        Property_Type.append('None')

    # Status
    try:
        sta = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[4]/div[2]/span/h4')
        status.append(sta.text)
    except:
        status.append('None')

    # Year Built
    try:
        Year = driver.find_element(By.XPATH, '//div[@class="outer-parameters-block"]/ul/li[10]/div[2]')
        year_text = Year.text.strip()
        
        if year_text.isdigit() and len(year_text) == 4:
            Year_Built.append(int(year_text))
        else:
            Year_Built.append('None')  # Append 'None' if not a valid year
    except:
        Year_Built.append('None')

    # Landmarks
    try:
        Landma = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[3]/div/div/span[2]/a[1]')
        Landmarks.append(Landma.text)
    except:
        Landmarks.append('None')

    # Balcony
    try:
        Balc = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[5]/div[2]/span/h4')
        if Balc.text.strip().upper() == 'NA':
            Balcony.append('None')
        else:
            Balcony.append(Balc.text.strip())
    except:
        Balcony.append('None')

    # Address
    try:
        Addr = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[1]/div/div/div[1]/div[2]/div[1]/h5')
        Address.append(Addr.text)
    except:
        Address.append('None')

    # City
    try:
        Cit = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[2]')
        city_text = Cit.text.split(',')[0]  # Split the text by comma and get the first part
        City.append(city_text)
    except:
        City.append('GreaterNoida')  # Default city assumed as 'Bengaluru'

    # State_code
    try:
        State_cod = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[2]')
        State_code_ = State_cod.text.split(',')[1].strip()  # Split the text by comma and get the second part
        State_code.append(State_code_)
    except:
        State_code.append('UP')  # Default state code assumed as 'KA'

    # Country
    try:
        Addr = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[2]')
        Country.append(Addr.text)
    except:
        Country.append('India')  # Default country assumed as 'India'

    # Floor
    try:
        fl = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[9]/div[3]/h5')
        if '/' in fl.text:
            floor.append(fl.text)
        else:
            floor.append('None')
    except:
        floor.append('None')

    # Transaction type
    try:
        trans = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        transaction_type.append(trans.text)
    except:
        transaction_type.append('None')

    # Furnished status
    try:
        furnish = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[7]/div[3]/h5')
        if 'Fully Furnished' in furnish.text or 'Semi' in furnish.text:
            furnished_status.append(furnish.text)
        else:
            furnished_status.append('None')
    except:
        furnished_status.append('None')

    # Facing
    try:
        fac = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[8]/div[3]/h5')
        fac_text = fac.text.lower()  # Convert text to lowercase for case insensitivity
    
        if 'north-east' in fac_text or 'northeast' in fac_text:
            facing.append('northeast')
        elif 'south-east' in fac_text or 'southeast' in fac_text:
            facing.append('southeast')
        elif 'south-west' in fac_text or 'southwest' in fac_text:
            facing.append('southwest')
        elif 'north-west' in fac_text or 'northwest' in fac_text:
            facing.append('northwest')
        elif 'north' in fac_text:
            facing.append('North')
        elif 'east' in fac_text:
            facing.append('East')
        elif 'south' in fac_text:
            facing.append('South')
        elif 'west' in fac_text:
            facing.append('West')
        else:
            facing.append('None')
    except:
        facing.append('None')

    # Car Parking
    try:
        Car_Par = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[2]/div[2]/div/section/div[2]/div[7]/div[2]/span/h4')
        Car_Parking.append(Car_Par.text)
    except:
        Car_Parking.append('None')

    # Type of ownership
    try:
        Type_of_owners = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[2]/div[3]/h5')
        Type_of_ownership.append(Type_of_owners.text)
    except:
        Type_of_ownership.append('None')

    # Booking amount
    try:
        booking = driver.find_element(By.XPATH, '//h1[@class="property-title-wrapper"]/span[10]')
        booking_amount.append(booking.text)
    except:
        booking_amount.append('None')

    
    # Buy total price
    CRORE_TO_THOUSAND = 10000000
    LAC_TO_THOUSAND = 100000

    def convert_price_to_thousands(price_text):
        if 'Crores' in price_text:
            price_value = float(price_text.replace('₹', '').replace('Crores', '').strip())
            price_in_thousands = price_value * CRORE_TO_THOUSAND
        elif 'Lacs' in price_text:
            price_value = float(price_text.replace('₹', '').replace('Lacs', '').strip())
            price_in_thousands = price_value * LAC_TO_THOUSAND
        else:
            price_in_thousands = float(price_text.replace('₹', '').strip())
        return price_in_thousands

    try:
        buy_t = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section/div/div[1]/div/div/section[1]/div/div[1]/div/div/div[1]/div[3]/div[1]/div/div[1]/span')
        price_text = buy_t.text
        price_in_thousands = convert_price_to_thousands(price_text)
        buy_total_price.append(f'{round(price_in_thousands)}') 
    except:
        buy_total_price.append('None')
    
# Quit WebDriver
driver.quit()

```
